```python
# 取事務所名稱、地址、電子郵件、電話、傳真
df_rowid = pd.read_excel("C:/Users/bbgog/land_rwoid_all.xlsx", header=None)
df_rowid.head()
```

```python
name_list = []
addr_list = []
email_list = []
tel_list = []
fax_list = []
```

```python
for i in range(8000,len(df_rowid)):
    rowid = df_rowid.iloc[i,0]
    url = 'https://resim.land.moi.gov.tw/Agent/iamqry_11c?rowid='+rowid
    df_detail_tmp = pd.read_html(url)[1]
    name = df_detail_tmp.iloc[0,0].split("(")[1].split(")")[0]
    url = 'https://resim.land.moi.gov.tw/Agent/iamqry_11d2?rowid='+rowid+'&acertname='+name
    response = rq.get(url) # 用 requests 的 get 方法把網頁抓下來
    html_doc = response.text # text 屬性就是 html 檔案
    soup = BeautifulSoup(response.text, "lxml") # 指定 lxml 作為解析器
    name = soup.find(id="acert_pract_org_name").get('value')
    addr = soup.find(id="acert_pract_org_addr").get('value')
    email = soup.find(id="acert_email").get('value')
    tel = soup.find(id="acert_phone1").get('value')
    fax = soup.find(id="acert_fax").get('value')
    
    name_list.append(name)
    addr_list.append(addr)
    email_list.append(email)
    tel_list.append(tel)
    fax_list.append(fax)
    
    print("has finished ", i)
```

```python
df_detail = pd.DataFrame({"name":name_list, "addr":addr_list, "email":email_list, "tel":tel_list, "fax":fax_list})
df_detail.head()
```
